```python
import numpy as np
import pandas as pd
```

```python
wine = pd.read_csv('/content/drive/MyDrive/WineQT.csv')
wine.head(10)
```

```python
wine['quality'] = [0 if x <= 6 else 1 for x in wine['quality']]
wine.head(10)
```

```python
features = wine.columns.difference(['quality'])
wine[features] = (wine[features] - wine[features].min()) / (wine[features].max() - wine[features].min())
wine.head(10)
```
